```python
import pandas as pd
from pandas.io.json import json_normalize
```

```python
companies = pd.read_json('../mongo/companies.json',lines=True)
```

```python
oficinas = companies['offices'][0]
oficinas
```

```python
df_oficinas = json_normalize(oficinas)
df_oficinas
```

```python
def obtenerUbicacionOficinas(lista):
    return json_normalize(lista)
```

```python
for i in range(1,len(companies['offices'])):
    df_oficinas = pd.concat([df_oficinas, obtenerUbicacionOficinas(companies['offices'][i])])
```

```python
len(df_oficinas)
```

```python
df_oficinas.to_csv('offices.csv')
```
